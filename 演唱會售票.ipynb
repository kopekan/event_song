### Cross Validation Split 

```python
def cv_split( k , data_size):
    cv = [0]    
    for i in range(k-1):
        cv.append(int((i+1)*np.floor(data_size/k)))
    cv.append(data_size)
    return cv
```

### Split Data into Basic & Overall Part 

```python
#如果不是放最原本的data進去，就不會分成basic, overall; out的basic, overall相關data就會是null            
def _split_data(train, test, target, divide, origin_shape):
    #target['total_quantity', 'total_sale', 'sale_prop_revenue', 'sale_prop_ticket']
    y_train = train[target].values.reshape((-1,1))
    y_test = test[target].values.reshape((-1,1))
    X_train = train
    X_test = test
    for i in ['total_quantity', 'total_sale', 'sale_prop_ticket', 'sale_prop_revenue']:
        try:
            X_train = X_train.drop(i, axis=1)
            X_test = X_test.drop(i, axis=1)
        except:
            print(i, 'is not exist.')

    if(divide):
        #index have to update
        basic_col = np.arange(list(X_train.columns).index('sale_month_12')+1) #僅基本資料
        overall_col = np.arange(list(X_train.columns).index('sale_month_12')+1,X_train.shape[1]) #僅非基本資料
        #split into basic and overall
        X_train_basic = X_train.iloc[:,basic_col]
        X_train_overall = X_train.iloc[:,overall_col]
        X_test_basic = X_test.iloc[:,basic_col]
        X_test_overall = X_test.iloc[:, overall_col]
    else :
         X_train_basic = X_train_overall = X_test_basic = X_test_overall = None  

#     print('basic:{} \t overall:{}'.format(len(basic_col), len(overall_col)))
#     print('X_train shape:{} \t y_train shape:{}'.format(X_train.shape, y_train.shape))
#     print('X_test shape:{} \t y_test shape:{}'.format(X_test.shape, y_test.shape)) 

    return X_train, X_test, X_train_basic, X_train_overall, X_test_basic, X_test_overall, y_train, y_test
```

### 2DNN+GBM

```python
def predict_2DNN_gbm_Model(X_train_basic, X_train_overall, X_test_basic, X_test_overall, y_train):    
    tf.keras.backend.clear_session()
    basic_input = tf.keras.Input(shape=X_train_basic.shape[1], name='basic_input')    #21 variables    
    overall_input = tf.keras.Input(shape=X_train_overall.shape[1], name='overall_input') #520 variables    
    #模型架構
    
    basic_model = layers.Dense(32,activation='sigmoid', name = 'basic_Dense_1')(basic_input)
    basic_model = layers.Dense(16,activation='sigmoid', name = 'basic_Dense_2')(basic_model) 
    basic_output = layers.Dense(8, name = 'basic_output')(basic_model) 
    
    basic_output = layers.Dense(32,activation='sigmoid', name = 'basic_Dense_1')(basic_input)
    
    overall_model = layers.Dense(256,activation='sigmoid', name = 'overall_Dense_1')(overall_input)        
    overall_model = layers.Dense(128,activation='sigmoid', name = 'overall_Dense_2')(overall_model)
    overall_model = layers.Dense(64,activation='sigmoid', name = 'overall_Dense_3')(overall_model)
 #   overall_model = layers.Dense(32,activation='sigmoid', name = 'overall_Dense_4')(overall_model)
    overall_output = layers.Dense(32, name = 'overall_output')(overall_model)        
    
    combine_model = layers.concatenate([basic_output, overall_output], axis = 1, name = 'combine_model')
    combine_output = layers.Dense(1,activation='sigmoid', name = 'combine_output')(combine_model)
    
    final_model = tf.keras.Model(inputs = [basic_input, overall_input], outputs = [combine_output])    
    final_model.compile(optimizer = 'adam'
            ,loss = tf.keras.losses.MeanSquaredError()
            ,metrics = [tf.keras.losses.MeanSquaredError()]
         )
    #Neural network fit
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10) #如果loss沒有下降，則再過10個epoch後停止
    final_model.fit([X_train_basic, X_train_overall], y_train, 
                    epochs=10000, batch_size=16, verbose=0
                    , validation_split = 0.2, callbacks=[early_stopping])
    #GBM
    gbm_input = tf.keras.Model(inputs = final_model.input, outputs = final_model.get_layer('combine_model').output)
    gbm_model = GradientBoostingRegressor(loss='ls', learning_rate=0.001, max_depth=3, n_estimators=1000) #ls=least square
    gbm_train_input = gbm_input.predict([X_train_basic, X_train_overall])
    gbm_model.fit(gbm_train_input, y_train)
    gbm_test_input = gbm_input.predict([X_test_basic, X_test_overall])
    predict = gbm_model.predict(gbm_test_input)
    if(np.max(y_train)<2):
        predict[predict>1]=1
    predict[predict<0]=0

    return final_model, predict.reshape((-1, 1))    
```

### DNN+CNN

```python
def predict_DNN_CNN_Model(X_train_basic, X_train_overall, X_test_basic, X_test_overall, y_train): 
    print('Using DNN+CNN Model')
    tf.keras.backend.clear_session()
    basic_input = tf.keras.Input(shape=X_train_basic.shape[1], name='basic_input')    #21 variables    
    overall_input   = tf.keras.Input(shape=(X_train_overall.shape[1], 1), name='overall_input') #520 variables
    #DNN - basic
    basic_model = layers.Dense(32,activation='sigmoid', name = 'basic_Dense_1')(basic_input)
    basic_model = layers.Dense(8,activation='sigmoid', name = 'basic_Dense_2')(basic_model) 
    basic_output = layers.Dense(1, name = 'basic_output')(basic_model) 
    #CNN - overall
    X_train_overall = X_train_overall.reshape(X_train_overall.shape[0], X_train_overall.shape[1],1)
    X_test_overall  = X_test_overall.reshape(X_test_overall.shape[0], X_test_overall.shape[1],1)

    overall_model = layers.Conv1D(filters=32, kernel_size=9, activation='sigmoid',padding='valid', name='overall_Conv_1')(overall_input)    
    overall_model = layers.MaxPooling1D(5)(overall_model)
    overall_model = layers.Conv1D(filters=16, kernel_size=5, activation='sigmoid', name = 'overall_Conv_2')(overall_model)
    overall_model = layers.MaxPooling1D(3)(overall_model)
    overall_model = layers.Conv1D(filters=8, kernel_size=3, activation='sigmoid', name = 'overall_Conv_3')(overall_model)
    overall_model = layers.MaxPooling1D(3)(overall_model)
    overall_model = layers.Flatten()(overall_model)
    overall_output = layers.Dense(1, name = 'overall_output')(overall_model)

    combine_model = layers.concatenate([basic_output, overall_output], axis = 1)
    combine_output = layers.Dense(1,activation='sigmoid', name = 'combine_output')(combine_model)

    final_model = tf.keras.Model(inputs = [basic_input, overall_input], outputs = [combine_output])    
    final_model.compile(optimizer = 'adam'
            ,loss = tf.keras.losses.MeanSquaredError()
            ,metrics = [tf.keras.losses.MeanSquaredError()]
         )
    
    #predict
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10) #如果loss沒有下降，則再過10個epoch後停止
    final_model.fit([X_train_basic, X_train_overall], y_train, epochs=10000, batch_size=16, verbose=0
                    , validation_split = 0.2, callbacks=[early_stopping])
    predict = final_model.predict([X_test_basic, X_test_overall])
    if(np.max(y_train)<2):
        predict[predict>1]=1
    predict[predict<0]=0

    return final_model, predict  
```

```python
def predict_GBM(X_train, X_test, y_train):
    gbm_model = GradientBoostingRegressor(loss='ls', learning_rate=0.1, max_depth=3, n_estimators=10000, n_iter_no_change=5, random_state = 314) #ls=least square
    gbm_model.fit(X_train, y_train)
    predict = gbm_model.predict(X_test)
    if(np.max(y_train)<2):
        predict[predict>1]=1
    predict[predict<0]=0
    return gbm_model, predict.reshape((-1,1))
```

```python
import xgboost as xgb
def predict_xgboost(X_train, X_test, y_train):
    dtrain = xgb.DMatrix(X_train, y_train)
    dtest = xgb.DMatrix(X_test)
    params = {
        'booster': 'gbtree',
        'objective': 'reg:linear',  
        'gamma': 0.1,                  # 用于控制是否后剪枝的参数,越大越保守，一般0.1、0.2这样子。
        'max_depth': 6,                # 构建树的深度，越大越容易过拟合
        'lambda': 2,                   # 控制模型复杂度的权重值的L2正则化项参数，参数越大，模型越不容易过拟合。
        'subsample': 0.8,              # 随机采样训练样本
        'colsample_bytree': 0.8,       # 生成树时进行的列采样
        'silent': 1,                   # 设置成1则没有运行信息输出，最好是设置为0.
        'eta': 0.01,                  # 如同学习率
        'seed': 314,
        'n_estimators':100,
        'nthread': 4                  # cpu 线程数
    }
    model = xgb.train(params.items(), dtrain, num_boost_round=10000, verbose_eval=False, early_stopping_rounds=5, evals = [(dtrain,'train')]) #num_rounds = 1000
    predict = model.predict(dtest)
    if(np.max(y_train)<2):
        predict[predict>1]=1
    predict[predict<0]=0
    return model, predict.reshape((-1,1))
```

```python
def predict_DNN_xgboost(X_train,new_feature_train, X_test, new_feature_test, y_train):    
    tf.keras.backend.clear_session()
    new_feature_train = np.array(new_feature_train.values)
    new_feature_test = np.array(new_feature_test.values)
#    basic_input = tf.keras.Input(shape=X_train_basic.shape[1], name='basic_input')    #21 variables    
#    overall_input = tf.keras.Input(shape=X_train_overall.shape[1], name='overall_input') #520 variables    
    mlp_input = tf.keras.Input(shape=X_train.shape[1], name='mlp_input')
    #模型架構    
    mlp_model = layers.Dense(128,activation='sigmoid', name = 'mlp_Dense_1')(mlp_input)
#    mlp_model = layers.Dense(128,activation='sigmoid', name = 'mlp_Dense_2')(mlp_model) 
#    mlp_model = layers.Dense(64,activation='sigmoid', name = 'mlp_Dense_3')(mlp_model)     
#    mlp_model = layers.Dense(32,activation='sigmoid', name = 'mlp_Dense_4')(mlp_model)
    mlp_model = layers.Dense(32, name = 'mlp_output')(mlp_model)        
    mlp_model = layers.Dense(1, name = 'mlp_output_temp')(mlp_model)    
    
    mlp = tf.keras.Model(inputs = mlp_input, outputs = mlp_model)    
    mlp.compile(optimizer = 'adam'
            ,loss = tf.keras.losses.MeanSquaredError()
            ,metrics = [tf.keras.losses.MeanSquaredError()]
         )    
    #MLP fit
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5) #如果loss沒有下降，則再過10個epoch後停止
    mlp.fit(X_train, y_train, 
            epochs=10000, batch_size=32, verbose=0
            , validation_split = 0.2, callbacks=[early_stopping])

    #xgboost
    gbm_input = tf.keras.Model(inputs = mlp.input, outputs = mlp.get_layer('mlp_output').output)
    gbm_train_input = gbm_input.predict(X_train)
    gbm_test_input = gbm_input.predict(X_test)
    
    gbm_train_input = np.concatenate((gbm_train_input, new_feature_train), axis = 1)
    gbm_test_input = np.concatenate((gbm_test_input, new_feature_test), axis = 1)
        
#    print('final gbm_train_input shape :',np.shape(gbm_train_input))
#    print('add new feature? ',new_feature_add)
    
    return predict_xgboost(gbm_train_input, gbm_test_input, y_train)    
```
